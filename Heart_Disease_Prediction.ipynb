#  *Importing the Libraries*

Numpy: To handle numerical data

Pandas: To work with tabular data

train_test_split: Split data into training and testing sets

LogisticRegression: Build a Logistic Regression model

accuracy_score: Evaluate model accuracy

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#  *Data Collection and Processing*

```python
# Loading the csv data to a Pandas Dataframe
heart_data=pd.read_csv("/dataset/heart-disease-prediction/heart_disease_data.csv")
```

```python
# View the first 5 rows of heart_data
heart_data.head()
```

**Target Variable:**

*1 --> Heart Issues*

*2 --> Healthy Heart*

```python
# Number of rows and columns in the datatset
heart_data.shape
```

```python
# Checking for Missing Values
heart_data.isnull().sum()
```

```python
# Statistical measures about the data
heart_data.describe()
```

```python
# Checking the distribution of Target Variable
heart_data['target'].value_counts()
```

# *Splitting the Features and Target*

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X.head()
```

```python
Y.head()
```

# *Splitting the Data into Training and Test data*

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# *Model Training*
We use a Logistic Regression Model since it is useful for Binary classification

```python
model = LogisticRegression()
```

```python
# Training our Logistic Regression Model
model.fit(X_train,Y_train)
```

# *Model Evaluation*

```python
# Using the metric accurac_score to check model performance
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data:', training_data_accuracy)
```

```python
# Using the metric accurac_score to check model performance
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data:', test_data_accuracy)
```

# *Building a Predictive System*

```python
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)
# Converting the input data into an array
input_data_as_np_array= np.asarray(input_data)
#Reshaping the numpy array to predict for only one instance
input_data_reshaped=input_data_as_np_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
if (prediction[0]==0):
  print('The person does not have any Heart Disease')
else:
  print('The person has a Heart Disease')
```
